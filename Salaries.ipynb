```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/salaries.csv')
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
print(df.isnull().sum())
```

```python
df['salary_in_usd'] = pd.to_numeric(df['salary_in_usd'])
```

```python
top_jobs = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(15)
top_jobs.plot(kind='bar',color='steelblue')
plt.title("Top 10 Job Titles by Average Salary")
plt.xlabel("Job Title")
plt.ylabel("Avg Salary (USD)")
plt.show()
```

```python
exp_level = df.groupby('experience_level')['salary_in_usd'].mean()
exp_level.plot(kind='line',marker = 'o',color = 'red')
plt.title("Average Salary by Experience Level")
plt.ylabel("Salary in USD")
plt.grid(True)
plt.show()
```

```python
country_df = df[df['company_location'].isin(['US','IN'])]
country_avg = country_df.groupby('company_location')['salary_in_usd'].mean()
country_avg.plot(kind='bar', color=['purple', 'green'])
plt.title("Average Salary: US vs India")
plt.ylabel("Salary (USD)")
plt.show()
```
